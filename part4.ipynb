```python
# A histogram is a type of bar plot that represents the distribution of numerical data. It is used to show the frequency (or probability density) of data points falling within certain ranges or bins
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = np.random.normal(loc=0,scale=1,size=1000)
sns.histplot(data,kde=True,color="blue",stat="density",bins=30)
plt.show()
```
