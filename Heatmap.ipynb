```python
#A heatmap is a graphical representation of data where individual values are represented by colors
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 12)  # 10x12 matrix with random values
sns.heatmap(data, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap Example')
plt.show()
```

```python
#Heatmap with Correlation Matrix

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a sample DataFrame
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [6, 5, 4, 3, 2]
})

corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#Customized Heatmap with Different Colors and Annotations
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(6, 6)  # 6x6 matrix with random values
sns.heatmap(data, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, linecolor='black')
plt.title('Customized Heatmap')
plt.show()
```

```python
#Heatmap with Clustering
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(10, 12)  # 10x12 matrix with random values
sns.clustermap(data, cmap='YlGnBu', annot=True)
plt.show()
```
