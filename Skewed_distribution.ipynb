```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# A positively skewed distribution (also known as right-skewed) is a type of distribution where the right tail (larger values) is longer or fatter than the left tail. This means that most of the data values are clustered on the left side, with a few larger values (outliers) on the right side. The mean of a positively skewed distribution is typically greater than the median
data1 = np.random.exponential(scale=1,size=1000)
sns.histplot(data1,kde=True,color="blue",stat="density",bins=30)
plt.show()

# A negatively skewed distribution (also known as left-skewed) is a distribution where the left tail (smaller values) is longer or fatter than the right tail. This means that most of the data values are clustered on the right side, with a few smaller values (outliers) on the left side. In a negatively skewed distribution, the mean is typically less than the median
data2 = -np.random.exponential(scale=1,size=1000)
sns.histplot(data2,kde=True,color="red",stat="density",bins=30)
plt.show()




sns.histplot(data1,kde=True,color="blue",stat="density",bins=30)
sns.histplot(data2,kde=True,color="red",stat="density",bins=30)
plt.show()

```
