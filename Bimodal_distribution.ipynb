```python

# A bimodal distribution refers to a distribution that has two distinct peaks or modes, rather than a single peak as in a normal distribution.


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data1 = np.random.normal(loc=-2, scale=1, size=1000)  
data2 = np.random.normal(loc=3, scale=1, size=1000)  

bimodal_data = np.concatenate([data1, data2])

sns.histplot(bimodal_data, kde=True, color="blue", stat="density", bins=30)
plt.title('Bimodal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()




```
